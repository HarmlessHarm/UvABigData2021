```python
# DATASET NEW YORK HOUSING

labels = ['LR - 1', 'LR - 2', 'LR - 3', 'RF - 1', 'RF - 2', 'RF - 3', 'NN - 1', 'NN - 2', 'NN - 3']

lr = [[0.050,  0.053, 0.052], [0.110, 0.118, 0.116], [0.158, 0.233, 0.187]]
rf = [[0.478, 0.726, 0.817], [2.183, 2.750, 2.914], [3.967, 5.708,  5.675]]
nn = [[20.766, 17.234, 18.172], [34.234, 40.188, 39.281], [57.391, 55.891, 59.516]]

def feature(n, m='time', d='NY'):
  lr = data[m][d]['lr']
  rf = data[m][d]['rf']
  nn = data[m][d]['nn']

  return [lr[0][n], lr[1][n], lr[2][n], rf[0][n], rf[1][n], rf[2][n], nn[0][n], nn[1][n], nn[2][n]]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars


for d in ["NY", 'RA', 'MU']:
  for m in ['time', 'acc']:
    fig, ax = plt.subplots()
    if m is 'time': 
      plt.yscale('log')
    ax.bar(x - width, feature(0, m, d), width, label='Feature 1')
    ax.bar(x, feature(1, m, d), width, label='Feature 2')
    ax.bar(x + width, feature(2, m, d), width, label='Feature 3')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel(m)
    ax.set_title(f"New York Housing Prices, {'Accuracy' if m == 'acc' else 'Time'}")
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    fig.tight_layout()
    plt.show()
```
